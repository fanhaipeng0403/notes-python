<p><span id="classmethod"></span></p>
<h4 id="classmethodfunctions">classmethod(functions)</h4>
<p>将传入的方法包装成类方法（类方法是指，类不需要实力化就可以直接调用的方法）。<br>
类方法的第一个参数必须为类（约定俗称用cls代表），实例方法的第一个参数必须为实例（约定俗称为self），这两种接收参数的方法叫做：隐式第一参数（implicit first argument）。静态方法（<a href="mailto:通过@staticmethod装饰的方法">通过@staticmethod装饰的方法</a>)不需要如上述两个方法的隐式参数。</p>
<div class="sourceCode"><pre class="sourceCode python"><code class="sourceCode python hljs"><span class="kw"><span class="hljs-class"><span class="hljs-keyword">class</span></span></span><span class="hljs-class"> <span class="hljs-title">C</span><span class="hljs-params">(</span></span><span class="bu"><span class="hljs-class"><span class="hljs-params">object</span></span></span><span class="hljs-class"><span class="hljs-params">)</span>:</span>
<span class="hljs-meta">    </span><span class="at"><span class="hljs-meta">@classmethod</span></span>
    <span class="kw"><span class="hljs-function"><span class="hljs-keyword">def</span></span></span><span class="hljs-function"> <span class="hljs-title">f</span><span class="hljs-params">(cls, arg1, arg2)</span>:</span>
        <span class="cf"><span class="hljs-keyword">pass</span></span></code></pre></div>
<p>注意：<a href="mailto:通过@classmethod修饰的方法">通过@classmethod修饰的方法</a>，为类方法。类可以直接调用（如C.f())；实例也可以直接调用（如C().f())，切记类方法中不能操作实例的属性。如果子类调用类方法，子类对象被传递为隐式的第一个参数（也就是cls为子类对象）。</p>
<p><span id="cmp"></span></p>
<h4 id="cmpx-y">cmp(x, y)</h4>
<p>比较两个对象x和y，当<code>x &lt; y</code>返回'-1'；<code>x &gt; y</code>返回'1';<code>x == y</code>返回0</p>
<div class="sourceCode"><pre class="sourceCode python"><code class="sourceCode python hljs">cmo(<span class="dv"><span class="hljs-number">1</span></span>, <span class="dv"><span class="hljs-number">3</span></span>) <span class="co"><span class="hljs-comment"># -1</span></span>
<span class="bu">cmp</span>(<span class="dv"><span class="hljs-number">3</span></span>, <span class="dv"><span class="hljs-number">1</span></span>) <span class="co"><span class="hljs-comment"># 1</span></span>
<span class="bu">cmp</span>(<span class="dv"><span class="hljs-number">1</span></span>, <span class="dv"><span class="hljs-number">1</span></span>) <span class="co"><span class="hljs-comment"># 0</span></span></code></pre></div>
<p>注意：<code>bool(-1)</code>结果为True</p>
<p><span id="compile"></span></p>
<h4 id="compilesource-filename-mode">compile(source, filename, mode)</h4>
<p>compile可以将字符串或者Unicode字符串编译成代码对象。代码对象可以通过<code>exec</code>或<code>eval</code>执行。</p>
<ul>
<li>参数source：字符串或者AST（Abstract Syntax Trees）对象。</li>
<li>参数 filename：代码文件名称，如果不是从文件读取代码则传递一些可辨认的值。</li>
<li><p>参数model：指定编译代码的种类。可以指定为 ‘exec’,’eval’,’single’。</p>
<div class="sourceCode"><pre class="sourceCode python"><code class="sourceCode python hljs">source <span class="op">=</span><span class="st"><span class="hljs-string">"if cmp(1, 1)==0: print 'yes'"</span></span>
test_compile <span class="op">=</span> <span class="bu">compile</span>(source, <span class="st"><span class="hljs-string">''</span></span>, <span class="st"><span class="hljs-string">'exec'</span></span>)
<span class="bu">exec</span>(test_compile)  <span class="co"><span class="hljs-comment">#yes</span></span></code></pre></div>
<p>注意：source是字符串，要注意引号和双引号，最后指定什么model，就用那种方法执行。compile返回的类型是<code>code</code>。</p></li>
</ul>

<h3 id="enumeratesequence-start0">enumerate(sequence, start=0)</h3>
<p>返回一个枚举对象。sequence必须是序列，迭代器iterator，或者支持迭代的对象：</p>
<blockquote>
<p>迭代器Iterators:<br>
迭代器仅是一容器对象，它实现了迭代器协议。它有两个基本方法：<br>
1）next方法<br>
返回容器的下一个元素<br>
2）__iter__方法<br>
返回迭代器自身</p>
</blockquote>
<div class="sourceCode"><pre class="sourceCode python"><code class="sourceCode python hljs">names <span class="op">=</span> [<span class="st"><span class="hljs-string">'Xuweihan'</span></span>, <span class="st"><span class="hljs-string">'James'</span></span>, <span class="st"><span class="hljs-string">'Kobe'</span></span>]

<span class="bu">list</span>(enumerate_data)  <span class="co"><span class="hljs-comment">#[(0, 'Xuweihan'), (1, 'James'), (2, 'Kobe')]</span></span>
