		<div class="postBody">
			<div id="cnblogs_post_body"><p>stat 系统调用时用来返回相关文件的系统状态信息的。</p>
<p>首先我们看一下stat中有哪些属性:</p>
<div class="cnblogs_Highlighter">
<pre class="brush:python;gutter:true;">&gt;&gt;&gt; import os
&gt;&gt;&gt; print os.stat("/root/python/zip.py")
(33188, 2033080, 26626L, 1, 0, 0, 864, 1297653596, 1275528102, 1292892895)
&gt;&gt;&gt; print os.stat("/root/python/zip.py").st_mode   #权限模式
33188
&gt;&gt;&gt; print os.stat("/root/python/zip.py").st_ino   #inode number
2033080
&gt;&gt;&gt; print os.stat("/root/python/zip.py").st_dev    #device
26626
&gt;&gt;&gt; print os.stat("/root/python/zip.py").st_nlink  #number of hard links
1
&gt;&gt;&gt; print os.stat("/root/python/zip.py").st_uid    #所有用户的user id
0
&gt;&gt;&gt; print os.stat("/root/python/zip.py").st_gid    #所有用户的group id
0
&gt;&gt;&gt; print os.stat("/root/python/zip.py").st_size  #文件的大小，以位为单位
864
&gt;&gt;&gt; print os.stat("/root/python/zip.py").st_atime  #文件最后访问时间
1297653596
&gt;&gt;&gt; print os.stat("/root/python/zip.py").st_mtime  #文件最后修改时间
1275528102
&gt;&gt;&gt; print os.stat("/root/python/zip.py").st_ctime  #文件创建时间
1292892895
</pre>
</div>
<p>&nbsp;正如你上面看到的，你可以直接访问到这些属性值。</p>
<p>好了，下面我来看看python中的stat模块，先看看自带的例子：</p>
<div class="cnblogs_code">
<pre><span style="color: #000000;">import os, sys
</span><span style="color: #0000ff;">from</span> stat import *<span style="color: #000000;">

def walktree(top, callback):
    </span><span style="color: #800000;">'''</span><span style="color: #800000;">recursively descend the directory tree rooted at top,</span>
       calling the callback function <span style="color: #0000ff;">for</span> each regular file<span style="color: #800000;">'''
</span>
    <span style="color: #0000ff;">for</span> f <span style="color: #0000ff;">in</span><span style="color: #000000;"> os.listdir(top):
        pathname </span>=<span style="color: #000000;"> os.path.join(top, f)
        mode </span>=<span style="color: #000000;"> os.stat(pathname).st_mode
        </span><span style="color: #0000ff;">if</span><span style="color: #000000;"> S_ISDIR(mode):
            # It</span><span style="color: #800000;">'</span><span style="color: #800000;">s a directory, recurse into it</span>
<span style="color: #000000;">            walktree(pathname, callback)
        elif S_ISREG(mode):
            # It</span><span style="color: #800000;">'</span><span style="color: #800000;">s a file, call the callback function</span>
<span style="color: #000000;">            callback(pathname)
        </span><span style="color: #0000ff;">else</span><span style="color: #000000;">:
            # Unknown file type, print a message
            print </span><span style="color: #800000;">'</span><span style="color: #800000;">Skipping %s</span><span style="color: #800000;">'</span> %<span style="color: #000000;"> pathname

def visitfile(file):
    print </span><span style="color: #800000;">'</span><span style="color: #800000;">visiting</span><span style="color: #800000;">'</span><span style="color: #000000;">, file

</span><span style="color: #0000ff;">if</span> __name__ == <span style="color: #800000;">'</span><span style="color: #800000;">__main__</span><span style="color: #800000;">'</span><span style="color: #000000;">:
    walktree(sys.argv[</span><span style="color: #800080;">1</span>], visitfile)</pre>
</div>
<p>可以这么理解，os.stat是将文件的相关属性读出来，然后用stat模块来处理，处理方式有多重，就要看看stat提供了什么了。</p>
<p>1.&nbsp; 可以对st_mode做相关的判断，如是否是目录，是否是文件，是否是管道等。</p>
<p>先看一下处理os.stat返回的st_mode结果的函数，就想上面的例子中的一样，这些函数可以做出判断：</p>
<div class="cnblogs_code">
<pre><span style="color: #0000ff;">if</span> stat.S_ISREG(mode):           <span style="color: #008000;">#</span><span style="color: #008000;">判断是否一般文件</span>
   <span style="color: #0000ff;">print</span> <span style="color: #800000;">'</span><span style="color: #800000;">Regular file.</span><span style="color: #800000;">'</span>
<span style="color: #0000ff;">elif</span> stat.S_ISLNK (mode):         <span style="color: #008000;">#</span><span style="color: #008000;">判断是否链接文件</span>
   <span style="color: #0000ff;">print</span> <span style="color: #800000;">'</span><span style="color: #800000;">Shortcut.</span><span style="color: #800000;">'</span>
<span style="color: #0000ff;">elif</span> stat.S_ISSOCK (mode):        <span style="color: #008000;">#</span><span style="color: #008000;">判断是否套接字文件    </span>
   <span style="color: #0000ff;">print</span> <span style="color: #800000;">'</span><span style="color: #800000;">Socket.</span><span style="color: #800000;">'</span>
<span style="color: #0000ff;">elif</span> stat.S_ISFIFO (mode):        <span style="color: #008000;">#</span><span style="color: #008000;">判断是否命名管道</span>
   <span style="color: #0000ff;">print</span> <span style="color: #800000;">'</span><span style="color: #800000;">Named pipe.</span><span style="color: #800000;">'</span>
<span style="color: #0000ff;">elif</span> stat.S_ISBLK (mode):         <span style="color: #008000;">#</span><span style="color: #008000;">判断是否块设备</span>
   <span style="color: #0000ff;">print</span> <span style="color: #800000;">'</span><span style="color: #800000;">Block special device.</span><span style="color: #800000;">'</span>
<span style="color: #0000ff;">elif</span> stat.S_ISCHR (mode):         <span style="color: #008000;">#</span><span style="color: #008000;">判断是否字符设置<span style="font-size: 12px; font-family: 宋体;"><br />　　<span style="color: #0000ff;">print</span> <span style="color: #800000;">'</span><span style="color: #800000;">Character special device.</span><span style="color: #800000;">'<br /></span></span></span></pre>
<pre><span style="color: #0000ff;">elif</span> stat.S_ISDIR (mode):         <span style="color: #008000;">#</span><span style="color: #008000;">判断是否目录<span style="font-size: 12px; font-family: 宋体;"><br />　　<span style="color: #0000ff;">print</span> <span style="color: #800000;">'</span><span style="color: #800000;">directory.</span><span style="color: #800000;">'<br />##额外的两个函数<br />stat.<tt class="descname">S_IMODE</tt> (mode): <span style="color: #008000;">#</span><span style="color: #008000;">返回文件权限的chmod格式<span style="font-size: 12px; font-family: 宋体;"><br />　　<span style="color: #0000ff;">print</span> <span style="color: #800000;">'</span><span style="color: #800000;">chmod format.</span><span style="color: #800000;">'</span></span></span></span></span></span></pre>
<pre></pre>
<pre><span style="color: #008000;"><span style="font-size: 12px; font-family: 宋体;"><span style="color: #800000;">stat.<tt class="descname">S_IFMT</tt> (mode): <span style="color: #008000;">#</span><span style="color: #008000;">返回文件的类型<span style="font-size: 12px; font-family: 宋体;"><br />　　<span style="color: #0000ff;">print</span> <span style="color: #800000;">'</span><span style="color: #800000;">type of fiel.</span><span style="color: #800000;">'</span></span></span></span></span></span></pre>
<pre><span style="color: #008000;">&nbsp;</span></pre>
</div>
<p>2.&nbsp;&nbsp; 还有一些是各种各样的标示符，这些标示符也可以在os.chmod中使用，下面附上这些标示符的说明：</p>
<div class="cnblogs_code">
<pre><span style="color: #000000;">    stat.S_ISUID: Set user ID on execution.                      不常用

    stat.S_ISGID: Set group ID on execution.                    不常用

    stat.S_ENFMT: Record locking enforced.                                          不常用

    stat.S_ISVTX: Save text image after execution.                                在执行之后保存文字和图片

    stat.S_IREAD: Read by owner.                                                           对于拥有者读的权限

    stat.S_IWRITE: Write by owner.                                                         对于拥有者写的权限

    stat.S_IEXEC: Execute by owner.                                                       对于拥有者执行的权限

    stat.S_IRWXU: Read, write, </span><span style="color: #0000ff;">and</span><span style="color: #000000;"> execute by owner.                          对于拥有者读写执行的权限

    stat.S_IRUSR: Read by owner.                                                            对于拥有者读的权限

    stat.S_IWUSR: Write by owner.                                                          对于拥有者写的权限

    stat.S_IXUSR: Execute by owner.                                                       对于拥有者执行的权限

    stat.S_IRWXG: Read, write, </span><span style="color: #0000ff;">and</span><span style="color: #000000;"> execute by group.                           　　　　　　对于同组的人读写执行的权限

    stat.S_IRGRP: Read by group.                                                             对于同组读的权限

    stat.S_IWGRP: Write by group.                                                           对于同组写的权限

    stat.S_IXGRP: Execute by group.                                                        对于同组执行的权限

    stat.S_IRWXO: Read, write, </span><span style="color: #0000ff;">and</span><span style="color: #000000;"> execute by others.                          对于其他组读写执行的权限

    stat.S_IROTH: Read by others.                                                           对于其他组读的权限

    stat.S_IWOTH: Write by others.                                                         对于其他组写的权限

    stat.S_IXOTH: Execute by others.                                                      对于其他组执行的权限</span></pre>
</div>

</span>&gt;&gt;&gt; st = os.stat(<span style="color: #800000;">'</span><span style="color: #800000;">sig.txt</span><span style="color: #800000;">'</span><span style="color: #000000;">)
